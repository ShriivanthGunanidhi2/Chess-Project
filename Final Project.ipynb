```python
# Average turns for each of the victory statuses?
```

```python
num_turns = list(df['turns'])
victory_status = list(df['victory_status'])
```

```python
status = {}
```

```python
for i in range(len(num_turns)):
    if victory_status[i] not in status.keys():
        status[victory_status[i]] = []
    status[victory_status[i]].append(num_turns[i])
```

```python
avg_turns_per_status = {}
```

```python
for move, turns in status.items():
    print(move)
    print(sum(turns) / len(turns))
```

```python
avg_turns_per_status
```

```python
# Any conclusions that can be drawn from the moves?
```

```python
conclusion = list(df['moves'])
```

```python
conclusion[0]
```

```python
e4_e5 = 0
just_e4 = 0
after_e4 = {'e5' : 0}
for move in conclusion:
    if 'e4 e5' in move[:5]:
        e4_e5 += 1
        after_e4['e5'] += 1
    elif 'e4' in move[:3]:
        just_e4 += 1
        if move[3:5] not in after_e4.keys():
            after_e4[move[3:5]] = 0
        after_e4[move[3:5]] += 1
```
